```python
def split_name(name):
    suffixes = [
    "Jr.", "Sr.", "II", "III", "IV", "V", "Esq.", "PhD", "MD", "DDS", "DVM",
    "CPA", "MBA", "MSc", "BSc", "DO", "OD", "RN", "JD", "LLD", "Eng.", "Ret.",
    "CFA", "CM", "OBE", "MBE", "CBE", "KBE", "DBE"
]      
    prefixes = [
    "Mr.", "Mrs.", "Ms.", "Miss", "Dr.", "Prof.", "Rev.", "Sir", "Madam", "Mx.",
    "Master", "Lady", "Dame", "Capt.", "Major", "Col.", "Gen.", "Lt.", "Hon.",
    "Pres.", "Gov.", "Sen.", "Amb.", "Arch.", "Sheikh", "Sultan", "Prince", "Princess"
]
    parts = name.split()
    if len(parts) == 2:
        return pd.Series([parts[0], parts[1], None, None])
    elif len(parts) >= 3:
        for prefix in prefixes:
            if parts[0] == prefix:
                parts.pop(0)
                break
        for suffix in suffixes:
            if parts[-1] == suffix:
                parts.pop()
                break
def count_parts(name):
    return len(name.split())
df['count'] = df['name'].apply(count_parts)
```
