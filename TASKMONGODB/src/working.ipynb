```python
import pandas as pd
from pymongo import MongoClient

# Connect
client = MongoClient("mongodb://localhost:27017/")
db = client["pythontraining"]
collection = db["project"]

# Pull documents into a cursor
cursor = collection.find({})   # {} = all documents

# Convert cursor to DataFrame
df = pd.json_normalize(cursor)

print(df.head(5))

```

```python
# Transformationsss
#Step1:
df1 = df[["project_id", "project_name", "client", "domain",
    "location", "project_manager", "start_date",
    "end_date", "status"]]
df1
```

```python
technologies = pd.DataFrame({
    "project_id":df["project_id"],
    "technologies":df["technologies"]
}).explode("technologies")
technologies
```

```python
technologies = pd.DataFrame({
"project_id":df["project_id"],
"technologies":df["technologies"]
}).explode("technologies")
project_tech_df = technologies.rename(columns={"technologies":"Technology"}).reset_index(drop=True)
project_tech_df
```

```python
status_map = {
    "In Progress": "Active",
    "Planned": "Pending",
    "Completed": "Done",
    
}
df1["status"] = df1["status"].map(status_map).fillna(df1["status"])
df1
```

```python
#Dates
df1["start_date"] = pd.to_datetime(df1["start_date"])
df1["end_date"] = pd.to_datetime(df1["end_date"])
df1 
```

```python
df1 = df1.drop_duplicates()
```
