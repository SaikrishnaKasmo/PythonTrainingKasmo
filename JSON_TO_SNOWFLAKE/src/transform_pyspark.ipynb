```python
from pyspark.sql import SparkSession

spark = (
    SparkSession
    .builder
    .appName("Spark Introduction")
    .master("local[*]")
    .getOrCreate()
)
```

```python
spark
```

```python
# Sample data
data = [
    (1, "Laptop", 1200.50),
    (2, "Phone", 800.00),
    (3, "Tablet", 300.75)
]

# Define schema (column names)
columns = ["product_id", "product_name", "price"]

# Create DataFrame
df = spark.createDataFrame(data, schema=columns)

df.show()
```
